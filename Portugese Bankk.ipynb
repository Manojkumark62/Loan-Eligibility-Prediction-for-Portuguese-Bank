## Hyperparameter Tuning

```python
from sklearn.model_selection import RandomizedSearchCV
```

### Random Forest Classifier tuning

```python
param_grid_rf = {
    'n_estimators':[100,200,300],
    'max_depth': [10,20,30,None],
    'min_samples_split': [2,5,10],
    'min_samples_leaf': [1,2,4],
    'class_weight': ['balanced',None]
}
```

```python
random_search_rf.fit(X_train_resampled,y_train_resampled)
```

```python
print('Best Random Forest Parameters:\n', random_search_rf.best_params_,'\n')
print('Best ROC AUC Score:\n', random_search_rf.best_score_)
```

# Report on Challenges faced

1. **Handling Outliers**
    - **Issue**: Oulier data points are seen in independent features of 'age', 'pdays', 'cons.conf.idx'. And I had not replaced those outliers.
    - **Solution**: Used tree based algorithms like Random Forest Classifier, XG Boost Classifier as they have less impact with the outliers.

2. **Handling Unknown Values**
    - **Issue**: Unknown values are seen in independent features of 'job', 'marital', 'education', 'default', 'housing', 'loan'.
    - **Solution**: I have considered it as a category as imputing them may change the data inturn afftecting the training.

3. **Handling Class Imbalance**
    - **Issue**: There is imbalance in the data of the target feature 'y'.
    - **Solution**: Fixed the imbalanced data using SMOTE (Synthetic Minority Over Sampling Technique) to match the minority class with the majority class in the training data.

4. **Dropping Duration**
    - **Issue**: Duration had data points which are very long at times.
    - **Solution**: Since duration of the call cannot be determined before contacting a customer, thought it is a redundant column and dropped the 'duration' independent feature.

5. **Multicolinearity**
    - **Issue**: There is a positive correlation between independent feature combinations of 'nr.employed (VS) euribor3m', 'euribor3m (VS) emp.var.rate' and 'nr.employed (VS) emp.var.rate'
    - **Solution**: Used tree based algorithms like Random Forest Classifier, XG Boost Classifier as the multicolinearity cannot affect the models performance.
